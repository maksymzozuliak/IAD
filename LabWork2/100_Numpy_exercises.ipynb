#### 78. Consider 2 sets of points P0,P1 describing lines (2d) and a point p, how to compute distance from p to each line i (P0[i],P1[i])? (★★★)

````python
def point_to_line_distances(p, P0, P1):
    p = np.array(p)
    P0 = np.array(P0)
    P1 = np.array(P1)

    v = P1 - P0
    w = p - P0

    t = np.sum(w * v, axis=1) / np.sum(v * v, axis=1)

    closest_points = P0 + t[:, np.newaxis] * v

    distances = np.linalg.norm(p - closest_points, axis=1)
    return distances

p = [1, 1]  # The point 'p'
P0 = np.array([[0, 0], [2, 2], [3, 4]])
P1 = np.array([[2, 0], [4, 2], [5, 5]])
distances = point_to_line_distances(p, P0, P1)
print(distances)
````

#### 80. Consider an arbitrary array, write a function that extract a subpart with a fixed shape and centered on a given element (pad with a `fill` value when necessary) (★★★)

````python
def extract_subpart(array, center, subpart_shape, fill_value=0):
    half_height, half_width = subpart_shape[0] // 2, subpart_shape[1] // 2

    start_row = max(0, center[0] - half_height)
    end_row = min(array.shape[0], center[0] + half_height + 1)
    start_col = max(0, center[1] - half_width)
    end_col = min(array.shape[1], center[1] + half_width + 1)

    subpart = np.full(subpart_shape, fill_value)

    overlap_start_row = half_height - (center[0] - start_row)
    overlap_end_row = overlap_start_row + (end_row - start_row)
    overlap_start_col = half_width - (center[1] - start_col)
    overlap_end_col = overlap_start_col + (end_col - start_col)

    subpart[overlap_start_row:overlap_end_row, overlap_start_col:overlap_end_col] = array[start_row:end_row, start_col:end_col]

    return subpart

array = np.arange(12).reshape(3,4)
center = (1, 2)
subpart_shape = (3, 3)

subpart = extract_subpart(array, center, subpart_shape, fill_value=0)
print(subpart)
````

#### 85. Create a 2D array subclass such that Z[i,j] == Z[j,i] (★★★)

````python
class Symetric(np.ndarray):
    def __setitem__(self, index, value):
        i,j = index
        super(Symetric, self).__setitem__((i,j), value)
        super(Symetric, self).__setitem__((j,i), value)

def symetric(Z):
    return np.asarray(Z + Z.T - np.diag(Z.diagonal())).view(Symetric)

temp1 = np.random.randint(0,10,(5,5))
S = symetric(temp1)
print(S)
````

````python
def cartesian(arrays):
    arrays = [np.asarray(a) for a in arrays]
    shape = (len(x) for x in arrays)

    ix = np.indices(shape, dtype=int)
    ix = ix.reshape(len(arrays), -1).T

    for n, arr in enumerate(arrays):
        ix[:, n] = arrays[n][ix[:, n]]

    return ix

print (cartesian(([1, 2], [4, 5], [6, 7])))
````
